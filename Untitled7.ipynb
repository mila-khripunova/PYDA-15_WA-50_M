```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="1")
with open('города.csv', 'r',encoding='utf-8') as g:
    with open('results_region.csv', 'w') as f:
        for line in g:
            try:
                location = geolocator.geocode(line.split(';')[1])
                f.write(location.address+','+str(location.latitude)+','+str(location.longitude)+','+line)
            except:
                f.write('')
g.close()
f.close()
#a = (location.address)
#print((location.latitude, location.longitude))
```
