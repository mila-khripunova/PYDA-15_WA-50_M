```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}


def p(documents_,directories_,psl_):  
    """Пользователь по команде "p" может узнать владельца документа по его номеру
Примеры работы:
1.
Введите команду:
p

Введите номер документа:
10006
Результат:
Владелец документа: Аристарх Павлов
2.
Введите команду:
p

Введите номер документа:
12345
Результат:
Документ не найден в базе"""
    
    a = 'Документ не найден в базе'
    if  psl_ == '':
        psl_ = str(input('Введите номер документа: '))
    for n,doc in enumerate(documents_):
        if doc['number'] == psl_:
            a = doc['name'] 
    return a,''

def s(documents_,directories_,psl_):
    """Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
Примеры работы:
1.
Введите команду:
s

Введите номер документа:
10006
Результат:
Документ хранится на полке: 2
2.
Введите команду:
p

Введите номер документа:
12345
Результат:
Документ не найден в базе"""
    a = 'Документ не найден в базе'
    if  psl_ == '':
        psl_ = str(input('Введите номер документа: '))
        
    for key,val in directories_.items():
        if psl_ in val:
            a = 'Документ хранится на полке: '+ key
    return a 


def l(documents_,directories_,psl_):
    """Пользователь по команде "l" может увидеть полную информацию по всем документам
Пример работы:
Введите команду:
l
Результат:
№: 2207 876234, тип: passport, владелец: Василий Гупкин, полка хранения: 1
№: 11-2, тип: invoice, владелец: Геннадий Покемонов, полка хранения: 1
№: 10006, тип: insurance, владелец: Аристарх Павлов, полка хранения: 2"""
    
    for n,doc in enumerate(documents_,1)
        print(n,doc['number'],'тип: ',doc['type'],'владелец: ',doc['name'],'полка хранения:',s(documents,directories,doc['number']))
    return '' 
            
def ads(documents_,directories_,psl_):
    """Пользователь по команде "ads" может добавить новую полку
Примеры работы:
1.
Введите команду:
ads

Введите номер полки:
10
Результат:
Полка добавлена. Текущий перечень полок: 1, 2, 3, 10.
2.
Введите команду:
ads

Введите номер полки:
1
Результат:
Такая полка уже существует. Текущий перечень полок: 1, 2, 3."""
    
    if  psl_ == '':
        psl_ = str(input('Введите номер полки: '))
        
    if psl_ in directories_:
        print('Такая полка уже существует. Текущий перечень полок:',', '.join(map(str, list(directories_.keys()))))
    else:
        directories_.setdefault(psl_),
        print('Полка добавлена. Текущий перечень полок:',', '.join(map(str, list(directories_.keys())))),
    return ''

def ds(documents_,directories_,psl_):
    """Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая)
Примеры работы:
1.
Введите команду:
ds

Введите номер полки:
3
Результат:
Полка удалена. Текущий перечень полок: 1, 2.
2.
Введите команду:
ds

Введите номер полки:
1
Результат:
На полке есть документа, удалите их перед удалением полки. Текущий перечень полок: 1, 2, 3.
3.
Введите команду:
ds

Введите номер полки:
4
Результат:
Такой полки не существует. Текущий перечень полок: 1, 2, 3."""
    if  psl_ == '':
        psl_ = str(input('Введите номер полки: '))
        
    if psl_ not in directories_:
        print('Такой полки не существует. Текущий перечень полок:',', '.join(map(str, list(directories_.keys()))))
    elif directories_[psl_]:
        print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:',', '.join(map(str, list(directories_.keys())))),
    else:
        directories_.pop(psl_),
        print('Полка удалена. Текущий перечень полок:',', '.join(map(str, list(directories_.keys()))))
    
    return ''


personal_functions = {'p': p,
                      's': s,
                      'l': l,
                      'ads':ads,
                      'ds':ds
 } 
user_input = '' 
while user_input !='q':  
    user_input = str(input('Введите команду: ')) 
    if user_input in personal_functions:
        function = personal_functions[user_input]
        print(function(documents,directories,'')) 
```
