## Задание  1

```python
import pandas as pd

data = pd.read_csv('ratings.csv')
data = data.merge(pd.read_csv('movies.csv'))
data.head()

def class_(param):
    if param['rating'] >= 4.5:
        return 'Hight'
    elif param['rating'] >= 2 and param['rating'] < 4.5:
        return 'Middle'
    elif param['rating'] < 2:
        return 'Low'
    

data['class'] = data.apply(class_, axis=1)
data.head(10)
```

## Задание 2

```python
import pandas as pd

keywords = pd.read_csv('keywords.csv')


geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

def geo(param):
    if param['keyword'] in ['петербург', 'псков', 'мурманск']:
        return 'Северо-Запад'
    if param['keyword'] in ['москва', 'тула', 'ярославль']:
        return 'Центр'   
    if param['keyword'] in ['владивосток', 'сахалин', 'хабаровск']:
        return 'Дальний Восток'
    return 'undefined'
    
    
keywords['region'] = keywords.apply(geo, axis=1)
keywords[keywords['region'].isin(geo_data.keys())].head(20)
```
